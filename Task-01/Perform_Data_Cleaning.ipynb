# ##Handling NaN

###Values considered â€œmissingâ€

NaN values can create inaccuracies in our estimations and calculations. There are two ways we can handle NaN:<br>
  1.we either remove them,<br>
  2.or we fill them.

# ###First method: Dropping NaN

dropna() : This will remove the row or column entries with NaN values.

```python
train_data.dropna(how="all")               #dropping when whole row is null
```

```python
train_data.shape
```

```python
train_data.dropna(inplace=True)## Remove NaN inside df only
train_data
```

fillna() : You can also fill NaN using a dict or Series that is alignable. The labels of the dict or index of the Series must match the columns of the frame you wish to fill.

Generally we fill the NaN values with the mean, but depending on the type of data, and your own analysis, you may decide to will NaN in some other way.

find 25% value and 75%value and match with statical data

```python
percent_25=train1_data["Age"].quantile(0.25)   # verify 25% is 21.0
percent_75=train1_data["Age"].quantile(0.75)   # verify 75% is 39.0
print(percent_25,percent_75)
```

find IQR(Range)=maximum(Q3)-minimum(Q1)

```python
IQR=percent_75-percent_25   
print(IQR)
```

find lowerlimit(minimum value ) upperlimit(Maximum value)

```python
lower_limit=percent_25-1.5*IQR
upper_limit=percent_75+1.5* IQR
print(lower_limit,upper_limit)
```

# outlier data

```python
outlier_data=train1_data[(train1_data.Age<lower_limit)|(train1_data.Age>upper_limit)]
outlier_data
```

```python
outlier_data.reset_index(drop = True, inplace = True)  ## Reset the indices
outlier_data
```

# No outlier data. data within cluster

```python
no_outlier=train1_data[(train1_data.Age>=lower_limit)&(train1_data.Age<upper_limit)]
no_outlier
```

```python
no_outlier.reset_index(drop = True, inplace = True)  ## Reset the indices
no_outlier
```

# Visualization of Age Column without Outliers

```python
sns.boxplot(x=no_outlier["Age"],showfliers=True)
```

# Capping : this is second method to remove the outlier

```python
new_df_cap=train1_data.copy()
new_df_cap['Age']=np.where(new_df_cap['Age']>upper_limit,
                           upper_limit,
                           np.where(new_df_cap['Age']<lower_limit,
                                    lower_limit,
                                    new_df_cap['Age']
                                   )
                          )
```

```python
new_df_cap.shape
```

```python
print('Before removing the outlier')
plt.figure(figsize=(5,5))
sns.displot(train1_data['Age'])
plt.show()
sns.boxplot(x=train1_data["Age"],showfliers=True)
plt.show()
print('After removing the outlier')
sns.displot(new_df_cap['Age'])
plt.show()
sns.boxplot(x=new_df_cap["Age"],showfliers=True)
plt.show()
```
