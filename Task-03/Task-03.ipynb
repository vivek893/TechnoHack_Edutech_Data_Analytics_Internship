# Histogram

Draw histogram and visualize the distribution of data in a dataset


```python
numeric_columns=data.select_dtypes(include=[float,int])
for column in numeric_columns.columns:
    bin=[1,2,3,4,5]
    plt.hist(data[column],bins=bin,alpha=0.7,label=column,edgecolor='yellow')
# Customize the plot
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histograms of Iris Dataset Columns")
plt.legend()
plt.show()
```

Draw Bar chart and visualize the distribution of data in a dataset: column name is species

```python
# Count the occurrences of each species
species_count=data['Species'].value_counts()
print(species_count)
# Create a bar chart
plt.figure(figsize=(8, 6))  # Set the figure size
# Plot the data
species_count.plot(kind='bar',color='red',edgecolor='black')
# Add labels and title
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Distribution of Iris Species')
plt.xticks(rotation=40)

# Show the plot
plt.show()
```

Draw histogram and visualize the distribution of data in a dataset

```python
# Create histograms for each numeric feature
plt.figure(figsize=(12, 8))  # Set the figure size

# Sepal Length Histogram
plt.subplot(2, 2, 1)
plt.hist(data['SepalLengthCm'], bins=20, color='skyblue')
plt.xlabel('SepalLengthCm')
plt.ylabel('Frequency')
plt.title('Distribution of SepalLengthCm')

# Sepal Width Histogram
plt.subplot(2, 2, 2)
plt.hist(data['SepalWidthCm'], bins=20, color='lightcoral')
plt.xlabel('SepalWidthCm')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Width')

# Petal Length Histogram
plt.subplot(2, 2, 3)
plt.hist(data['PetalLengthCm'], bins=20, color='lightgreen')
plt.xlabel('PetalLengthCm')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Length')

# Petal Width Histogram
plt.subplot(2, 2, 4)
plt.hist(data['PetalWidthCm'], bins=20, color='gold')
plt.xlabel('PetalWidthCm')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Width')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
